```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

# Set the mean and covariance matrix for the distribution
mean = np.array([0, 2])
covariance = np.array([[0.3, -1], [-1, 5]])

# Generate 200 random samples from the distribution
samples = np.random.multivariate_normal(mean, covariance, 200)

# Create a meshgrid for the bivariate normal distribution
x, y = np.meshgrid(np.linspace(-1, 1, 100), np.linspace(-1, 10, 100))
pos = np.dstack((x, y))

# Calculate the PDF for the distribution
pdf = multivariate_normal.pdf(pos, mean, covariance)

# Plot the contours of the distribution
plt.contour(x, y, pdf, cmap='Blues', levels=10, alpha=0.5)

# Create a scatter plot for the simulated points
plt.scatter(samples[:, 0], samples[:, 1], label='Simulated Points', color='red')

# Plot the mean with a marker
plt.scatter(mean[0], mean[1], marker='X', color='green', label='Mean')

# Add text annotation for the mean
plt.annotate('Mean', xy=(mean[0], mean[1]), xytext=(mean[0] + 0.5, mean[1] + 0.5),
             arrowprops=dict(facecolor='black', shrink=0.05))

# Set labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Bivariate Gaussian Distribution with Simulated Points')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
